```python
# Import necessary libraries
import pandas as pd
import numpy as np

# Load the dataset
file_path = 'Gold_reserves_tonnes.csv'
data = pd.read_csv(file_path)


# Display the first few rows of the cleaned dataset
print(data.head())
```

```python
# Handle missing values
data.fillna(0, inplace=True)  # Replace missing values with 0

# Ensure data types are consistent
data['Average gold reserves'] = data['Average gold reserves'].astype(float)

# Detect and remove outliers using the IQR method
Q1 = data['Average gold reserves'].quantile(0.25)
Q3 = data['Average gold reserves'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
data = data[(data['Average gold reserves'] >= lower_bound) & (data['Average gold reserves'] <= upper_bound)]
```

```python
# Preprocess the data for neural network training
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer

# Assuming 'Average gold reserves' is the target variable
X = data.drop(columns=['Average gold reserves'])
y = data['Average gold reserves']

# Identify numeric and categorical columns
numeric_features = X.select_dtypes(include=['int64', 'float64']).columns
categorical_features = X.select_dtypes(include=['object']).columns

# Apply transformations: scale numeric features and encode categorical features
preprocessor = ColumnTransformer(
    transformers=[
        ('num', MinMaxScaler(), numeric_features),
        ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_features)
    ]
)

# Transform the features
X_processed = preprocessor.fit_transform(X)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_processed, y, test_size=0.2, random_state=42)
```

```python
# Build and train a neural network model using scikit-learn
from sklearn.neural_network import MLPRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Define the model
model = MLPRegressor(hidden_layer_sizes=(64, 32), activation='relu', solver='adam', max_iter=500, random_state=42)

# Train the model
model.fit(X_train, y_train)

# Evaluate the model
y_pred = model.predict(X_test)
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f'Mean Absolute Error on test data: {mae}')
print(f'Mean Squared Error on test data: {mse}')
print(f'R-squared on test data: {r2}')
```
