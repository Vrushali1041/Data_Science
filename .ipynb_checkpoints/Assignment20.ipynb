```python
# Import necessary libraries
import pandas as pd
import numpy as np

# Load the dataset
file_path = 'Gold_ demand_quarterly.csv'
data = pd.read_csv(file_path)

# Display the first few rows of the dataset
print("Initial Dataset:")
print(data.head())

# Handle missing values (if any)
data.fillna(0, inplace=True)

# Convert columns to appropriate data types (if necessary)
for col in data.columns[1:]:
    data[col] = pd.to_numeric(data[col], errors='coerce')

# Display summary statistics
print("\nSummary Statistics:")
print(data.describe())

# Save the preprocessed dataset (optional)
preprocessed_file_path = 'E:\Gold_demand_preprocessed.csv'
data.to_csv(preprocessed_file_path, index=False)
print(f"\nPreprocessed dataset saved to {preprocessed_file_path}")
```

```python
from scipy.stats import zscore

# Calculate Z-scores for numerical columns
z_scores = data.iloc[:, 1:].apply(zscore)

# Define a threshold for identifying outliers
threshold = 3

# Identify rows with outliers
outliers = (z_scores.abs() > threshold).any(axis=1)

# Remove rows with outliers
data_cleaned = data[~outliers]

print("Data after removing outliers:")
print(data_cleaned)

# Optionally, save the cleaned dataset
cleaned_file_path = 'e:\\Parth Mca 2nd sem\\Data Science\\Data Science Assignment\\Gold_demand_cleaned.csv'
data_cleaned.to_csv(cleaned_file_path, index=False)
print(f"\nCleaned dataset saved to {cleaned_file_path}")
```

```python
from sklearn.cluster import KMeans

import matplotlib.pyplot as plt

# Select numerical columns for clustering (excluding 'Countries')
numerical_data = data_cleaned.iloc[:, 1:]

# Determine the optimal number of clusters using the Elbow Method
inertia = []
k_range = range(1, 11)
for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(numerical_data)
    inertia.append(kmeans.inertia_)

# Plot the Elbow Curve
plt.figure(figsize=(8, 5))
plt.plot(k_range, inertia, marker='o')
plt.title('Elbow Method for Optimal K')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.grid()
plt.show()

# Perform K-means clustering with the chosen number of clusters (e.g., k=3)
kmeans = KMeans(n_clusters=3, random_state=42)
data_cleaned['Cluster'] = kmeans.fit_predict(numerical_data)

# Display the clustered data
print("Clustered Data:")
print(data_cleaned[['Countries', 'Cluster']])
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error

# Assuming the true values and predicted values are available
# Replace `true_values` and `predicted_values` with actual data
true_values = numerical_data.values.flatten()
predicted_values = kmeans.cluster_centers_[data_cleaned['Cluster']].flatten()

# Calculate MAE and MSE
mae = mean_absolute_error(true_values, predicted_values)
mse = mean_squared_error(true_values, predicted_values)

print(f"Mean Absolute Error (MAE): {mae}")
print(f"Mean Squared Error (MSE): {mse}")
```
